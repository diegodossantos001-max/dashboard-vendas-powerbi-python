```python
import pandas as pd
```

```python
df = pd.read_csv("../dados/vendas_brutas.csv", sep=";")
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
df["data_venda"] = pd.to_datetime(
    df["data_venda"],
    dayfirst=True,
    errors="coerce"
)

df["ano"] = df["data_venda"].dt.year
df["mes"] = df["data_venda"].dt.month

meses_ptbr = {
    1: "Janeiro",
    2: "Fevereiro",
    3: "Março",
    4: "Abril",
    5: "Maio",
    6: "Junho",
    7: "Julho",
    8: "Agosto",
    9: "Setembro",
    10: "Outubro",
    11: "Novembro",
    12: "Dezembro"
}

df["nome_mes"] = df["mes"].map(meses_ptbr)

df.head()
```

```python
df["data_venda"].isnull().sum()
```

```python
df["faturamento"] = df["quantidade"] * df["preco_unitario"]

df.head()
```

```python
df["vendedor"] = df["vendedor"].str.title()
df["produto"] = df["produto"].str.title()
df["categoria"] = df["categoria"].str.title()
df["cidade"] = df["cidade"].str.title()
df["forma_pagamento"] = df["forma_pagamento"].str.title()

df.head()
```

```python
df = df.drop_duplicates()

df.shape
```

```python
df.groupby("produto")["faturamento"].sum().sort_values(ascending=False)
```

```python
df.to_csv("../dados/vendas_tratadas.csv", sep=";", index=False, encoding="utf-8-sig")

print("Arquivo vendas_tratadas.csv gerado com sucesso!")
```
